```python
#WAP to calculate the area of a circle

def area_of_circle(r):
    return cons.pi * r**2

radius = int(input("Enter the radius: "))
area = area_of_circle(radius)

print(f"The area of circle with radius {radius} is: {round(area,2)}")
```

```python
#WAP to convert kilometer into meter

def km_to_m(km):
    return km * cons.kilo

km = float(input("Enter the km: "))
meter = km_to_m(km)

print(f"{km} km is equivalent to {meter} meters.")
```

```python
#Program to convert celsius to kelvin

cel = float(input("Enter the temp. in Celsius: "))

kelvin = cel + cons.zero_Celsius

print(f"{cel} C is equivalent to {kelvin} K.")
```
