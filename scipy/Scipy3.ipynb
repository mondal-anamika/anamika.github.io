```python
import numpy as np
from scipy.linalg import solve
```

```python
A = np.array([[2, 3],[1, -4]])
b = np.array([8, 2])

#solve the system of linear equations Ax = b
x_sol = solve(A, b)

#printing the solution
print("Solution for Ax=b", x_sol)
```

```python
from scipy import datasets
from scipy import ndimage
import matplotlib.pyplot as plt

#read the racoon face
image = datasets.face()

#The original image
plt.imshow(image)
plt.show()
```

```python
#add gaussian filter to the image
gaussian_image = ndimage.gaussian_filter(image, sigma=5)

plt.imshow(gaussian_image)
plt.show()
```

```python
#adding median filter to the image
median_image = ndimage.median_filter(image, size=4)

plt.imshow(median_image)
plt.show()
```

```python
#laplace filter
laplace_image = ndimage.laplace(image)

plt.imshow(laplace_image)
plt.show()
```
